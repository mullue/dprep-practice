## 4. Join, Merge, Concat

### Join
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html?highlight=join#pandas.DataFrame.join

`````python
df = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', 'K4', 'K5'],'A': ['A0', 'A1', 'A2', 'A3', 'A4', 'A5']})
df
`````

`````python
other = pd.DataFrame({'key': ['K0', 'K1', 'K2'],'B': ['B0', 'B1', 'B2']})
other
`````

- 공통의 컬럼을 기반으로 조인

`````python
df.join(other, lsuffix='_caller', rsuffix='_other')
`````

- index 컬럼을 기반으로 조인

`````python
df.set_index('key').join(other.set_index('key'))
`````

- on 키워드를 이용한 조인 (이 경우, 조인대상 데이터프레임의 index 를 조인컬럼으로 설정해야 함)

`````python
df.join(other.set_index('key'), on='key')
`````

### Merge
- https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.merge.html?highlight=merge#pandas.DataFrame.merge

`````python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'], 'value': [1, 2, 3, 5]})
df1
`````

`````python
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'], 'value': [5, 6, 7, 8]})
df2
`````

`````python
df1.merge(df2, left_on='lkey', right_on='rkey')
`````

`````python
df1.merge(df2, left_on='lkey', right_on='rkey',  suffixes=('_left', '_right'))
`````

- Merge를 이용한 조인

`````python
df1 = pd.DataFrame({'a': ['foo', 'bar'], 'b': [1, 2]})
df1
`````

`````python
df2 = pd.DataFrame({'a': ['foo', 'baz'], 'c': [3, 4]})
df2
`````

`````python
df1.merge(df2, how='inner', on='a')
`````

`````python
df1.merge(df2, how='left', on='a')
`````

`````python
df1.merge(df2, how='right', on='a')
`````

## 5. Pivot, Melt

### Pivot
- https://pandas.pydata.org/docs/reference/api/pandas.pivot.html

`````python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two', 'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
`````

`````python
df.pivot(index='foo', columns='bar')
`````

`````python
df.pivot(index='foo', columns='bar', values='baz')
`````

`````python
df.pivot(index='foo', columns='bar')['zoo']
`````

### Pivot_table
- https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html

`````python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo", "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two", "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small", "small", "large", "small", "small", "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
`````

- pivot_talbe의 경우 aggfunc를 사용할 수 있음

`````python
table = pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'], aggfunc=np.sum)
table
`````

`````python
table = pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'], aggfunc=np.sum, fill_value=0)
table
`````

`````python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'], aggfunc={'D': np.mean, 'E': np.mean})
table
`````

`````python
table = pd.pivot_table(df, values=['D', 'E'], index=['A', 'C'], aggfunc={'D': np.mean, 'E': [min, max, np.mean]})
table
`````

### Melt

`````python
pd.melt(df, id_vars=['A'], value_vars=['B'])
`````

`````python
pd.melt(df, id_vars=['A'], value_vars=['B', 'C'])
`````

`````python
pd.melt(df, id_vars=['A'], value_vars=['B'], var_name='myVarname', value_name='myValname')
`````

### Crosstab
- Confusion Matrix 생성 등에 사용됩니다.
- https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html

`````python
a = np.array(["foo",  "foo",  "foo",   "foo",  "bar",  "bar",   "bar",   "bar",  "foo",   "foo",   "foo"], dtype=object)
b = np.array(["one",  "one",  "one",   "two",  "one",  "one",   "one",   "two",  "two",   "two",   "one"], dtype=object)
c = np.array(["dull", "dull", "shiny", "dull", "dull", "shiny", "shiny", "dull", "shiny", "shiny", "shiny"], dtype=object)
print(a)
print(b)
print(c)
`````

- Series 타입의 데이터를 입력받아 빈도표를 생성할 수 있다.

`````python
pd.crosstab(a, [b, c], rownames=['a'], colnames=['b', 'c'])
`````

`````python
foo = pd.Categorical(['a', 'b'], categories=['a', 'b', 'c'])
bar = pd.Categorical(['d', 'e'], categories=['d', 'e', 'f'])
print(foo)
print(bar)
`````

`````python
type(foo)
`````

`````python
pd.crosstab(foo, bar)
`````

`````python
pd.crosstab(foo, bar, dropna=False)
`````
